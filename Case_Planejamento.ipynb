```python
#função pra deixar as tabelas na mesma linha (lado a lado)
def display_side_by_side(dataframes:list,titles:list):
    html_str=''
    for df,title in zip(dataframes, titles):
        html_str+=f'<span>{title}:</span>'
        html_str+=df.to_html().replace('table','table style="display:inline"')
        html_str+="&nbsp"*10
    display(HTML(html_str))
```

1 - Na guia "Base de Dados" preencha a coluna "Empresa" utilizando a fórmula PROCV com a guia "Auxiliar"

```python
df = df.rename(columns={'Nome Operador': 'Operador'})
```

```python
df.columns = map(str.lower, df.columns)
df_Auxiliar_OE.columns = map(str.lower, df_Auxiliar_OE.columns)
```

```python
df_Auxiliar_OE['operador'] = df_Auxiliar_OE['operador'].str.lower()
```

```python
df = pd.merge(df, df_Auxiliar_OE[['operador', 'empresa']], on='operador', how='left')
df.head(4)
```

2 - Na guia "Base de Dados" preencha a coluna "Canal de Atendimento" utilizando a formula PROCV com a guia "Auxiliar"

```python
df_Auxiliar_OC.columns = map(str.lower, df_Auxiliar_OC.columns)
```

```python
df_Auxiliar_OC['operador'] = df_Auxiliar_OC['operador'].str.lower()
```

```python
df = pd.merge(df, df_Auxiliar_OC[['operador', 'canal de atendimento']], on='operador', how='left')
df.head(4)
```

3 - Na guia "Base de Dados" preencha a coluna "Desconto" o percentual de desconto aplicado no campo "Valor da Venda", o valor original do produto está na guia "Auxiliar"

```python
df_Preco_Original.columns = map(str.lower, df_Preco_Original.columns)
```

```python
df_Preco_Original['valor do plano'] = df_Preco_Original['valor do plano'].replace('R\$', '', regex=True)
df_Preco_Original['valor do plano'] = df_Preco_Original['valor do plano'].replace(' ', '', regex=True)
df_Preco_Original['valor do plano'] = df_Preco_Original['valor do plano'].replace(',', '.', regex=True)
df_Preco_Original['valor do plano'] = df_Preco_Original['valor do plano'].astype(float)
df_Preco_Original
```

```python
df['valor da venda'] = df['valor da venda'].astype(float)
```

```python
df = pd.merge(df, df_Preco_Original, on=['produto'])
df['desconto'] = (1 - (df['valor da venda'] / df['valor do plano'])) * 100
colunas_para_apagar = ['valor do plano']
df = df.drop(colunas_para_apagar, axis=1) 
df.head(4)
```

4 - Na guia "Base de Dados" preencha a coluna "Status do Pagamento" informando o status da venda (Pago/Em aberto) utilizando a formula SE

```python
def status_pagamento(row):
    if pd.isna(row["data pagamento"]):
        return "Em aberto"
    else:
        return "Pago"

df["status do pagamento"] = df.apply(status_pagamento, axis=1)
#visualiza 
df.head(4)
```

5 - Na guia "Base de Dados" preencha a coluna "Data da venda (dia da semana)" informando a dia da semana ex. seg, ter ...

```python
df["data da venda (dia da semana)"] = df["data da venda"].dt.day_name() 
df.head(4)
```

```python
dias_da_semana = {
    "Monday": "Segunda-feira",
    "Tuesday": "Terça-feira",
    "Wednesday": "Quarta-feira",
    "Thursday": "Quinta-feira",
    "Friday": "Sexta-feira",
    "Saturday": "Sábado",
    "Sunday": "Domingo"
}

df["data da venda (dia da semana)"] = df["data da venda (dia da semana)"].map(dias_da_semana)
df.head(4)
```

6 - Na guia "Base de Dados" coloque o nome do Operador em letra maiuscula 

```python
df['operador'] = df['operador'].str.upper()
df.head(4)
```

7 -Junte-se ao nome do Operador com a Empresa. EX: "Tiago + Empresa 1"

```python
df['nome + empresa'] = df.apply(lambda x: x['operador'] + ' ' + x['empresa'], axis=1)
df.head(4)
```

Extra

```python
df['data da venda'] = pd.to_datetime(df['data da venda'])
df['data pagamento'] = pd.to_datetime(df['data pagamento'])
df['dias_de_atraso'] = (df['data pagamento'] - df['data da venda']).dt.days.fillna(-1).astype(int)
df.head()
```

## Exploratory Data Analysis (EDA) + Gráficos 

```python
df['dia_mes'] = df['data da venda'].dt.strftime('%d/%m')
```

```python
status_grouped = df.groupby(['dia_mes' , 'status do pagamento']).size().sort_values(ascending=False).reset_index()
status_grouped
status_grouped = status_grouped.rename(columns={0: 'Contagem'})
status_grouped = status_grouped.sort_values('dia_mes')

fig = px.bar(status_grouped, x='dia_mes', y='Contagem', color='status do pagamento',
             title='Contagem de vendas por data e status de pagamento',
             labels={'Contagem': 'Contagem de vendas', 'Status do Pagamento': 'Status de pagamento'})
fig.update_layout(barmode='stack')
fig.update_layout(xaxis=dict(range=[min(status_grouped['dia_mes']), max(status_grouped['dia_mes'])]))
fig.update_layout(template='plotly_dark')
fig.show()
```

```python
fig = px.line(status_grouped, x='dia_mes', y='Contagem', color='status do pagamento', title='status do pagamento x Mês - Dia' )
fig.update_xaxes(title_text='Mês - 2022')
fig.update_yaxes(title_text='Contagem de vendas')
fig.update_layout(template='plotly_dark')
fig.update_layout(hovermode="x unified")
fig.update_traces(hoverinfo='text+name', mode='lines+markers')
fig.show()
```

```python
Semana_groupby = df.groupby(['data da venda (dia da semana)'])['valor da venda'].mean().round(2).reset_index()

def weekday_to_number(day):
    if day == 'Segunda-feira':
        return 1
    elif day == 'Terça-feira':
        return 2
    elif day == 'Quarta-feira':
        return 3
    elif day == 'Quinta-feira':
        return 4
    elif day == 'Sexta-feira':
        return 5
    else:
        return 0 # caso o valor da coluna seja inválido
    
Semana_groupby['dia_da_semana'] = Semana_groupby['data da venda (dia da semana)'].apply(weekday_to_number)
Semana_groupby = Semana_groupby.sort_values(by='dia_da_semana')
Semana_groupby = Semana_groupby.drop('dia_da_semana', axis=1)
Semana_groupby
```

```python
fig = px.bar(Semana_groupby, x='data da venda (dia da semana)', y='valor da venda', title='Valor médio da venda por dia da semana')
fig.add_scatter(x=Semana_groupby['data da venda (dia da semana)'], y=Semana_groupby['valor da venda'], mode='lines+markers', name='valor da venda')
fig.update_layout(template='plotly_dark')
fig.show()
```

```python
grouped = df.groupby(['produto', 'forma de pagamento'])['valor da venda'].sum().reset_index()
fig = px.bar(grouped, x='produto', y='valor da venda', color='forma de pagamento',
             title='Valor de vendas por produto e forma de pagamento', barmode='group')
fig.update_layout(template='plotly_dark')
fig.show()
```

```python
discount_counts = df.groupby('desconto')['valor da venda'].count().reset_index(name='Quantidade')
discount_counts['desconto'] = discount_counts['desconto'].astype(int)
discount_counts = discount_counts.sort_values('desconto')
fig = px.pie(discount_counts, values='Quantidade', names='desconto', title='Quantidade de vendas por desconto' , hole=0.5)
fig.update_layout(template='plotly_dark')
fig.show()
```

```python
total_vendas = df.groupby('operador')['valor da venda'].sum()
operadores_ordenados = total_vendas.sort_values(ascending=False).index.tolist()


fig = px.bar(df, x='operador', y='valor da venda', color='operador',
             title='Ranking de Operadoras por Quantidade de Vendas',
             category_orders={"operador": operadores_ordenados})

fig.add_scatter(x=operadores_ordenados,
                y=total_vendas[operadores_ordenados].values,
                mode='lines+markers', name='Valor de Vendas')

fig.update_layout(xaxis_title='Operador', yaxis_title='Valor de Vendas',
                  template='plotly_dark')

fig.show()
```

```python
df_sem_menos_um = df.query("`dias_de_atraso` > 0").reset_index(drop=True)
media_atraso_produto = df_sem_menos_um.groupby('produto')['dias_de_atraso'].agg(['mean', 'max', 'min']).reset_index()
media_atraso_produto['mean'] = media_atraso_produto['mean'].round(2)
media_atraso_produto
```

```python
produtos = ['Premium Anual', 'Premium Mensal']
medias = [3.46, 5.24]
maximos = [22, 10]
minimos = [1, 1]

fig1 = go.Figure(data=[go.Pie(labels=produtos, values=medias, hole=0.5)])
fig1.update_layout(title='Média de Dias de Atraso por Produto')  

fig2 = go.Figure()
fig2.add_trace(go.Bar(x=produtos, y=medias, name='Média'))
fig2.add_trace(go.Bar(x=produtos, y=maximos, name='Máximo'))
fig2.add_trace(go.Bar(x=produtos, y=minimos, name='Mínimo'))
fig2.update_layout(title='Estatísticas de Dias de Atraso por Produto', barmode='group')

fig1.update_layout(template='plotly_dark')
fig2.update_layout(template='plotly_dark')

fig1.show()
fig2.show()
```

```python
df_sem_menos_um = df.query("`dias_de_atraso` > 0").reset_index(drop=True)
media_atraso_produto = df_sem_menos_um.groupby(['produto', 'forma de pagamento'])['dias_de_atraso'].agg(['mean', 'max', 'min']).reset_index()
media_atraso_produto['mean'] = media_atraso_produto['mean'].round(2)
media_atraso_produto
```

```python
df_sem_menos_um = df.query("`dias_de_atraso` > 0").reset_index(drop=True)
media_atraso_produto_forma_pagamento = df_sem_menos_um.groupby(['produto', 'forma de pagamento'])['dias_de_atraso'].agg(['mean', 'max', 'min']).reset_index()
media_atraso_produto_forma_pagamento['mean'] = media_atraso_produto_forma_pagamento['mean'].round(2)
media_atraso_produto_forma_pagamento
fig1 = go.Figure(data=[go.Pie(labels=media_atraso_produto_forma_pagamento['produto'], 
                              values=media_atraso_produto_forma_pagamento['mean'], 
                              hole=0.5, 
                              text=media_atraso_produto_forma_pagamento['forma de pagamento'],
                              textposition='inside'
                             )])
fig1.update_layout(title='Média de Dias de Atraso por Produto e Forma de Pagamento')

produtos = media_atraso_produto_forma_pagamento['produto'].unique()
fig2 = go.Figure()
for forma_pagamento in media_atraso_produto_forma_pagamento['forma de pagamento'].unique():
    df_forma_pagamento = media_atraso_produto_forma_pagamento[media_atraso_produto_forma_pagamento['forma de pagamento'] == forma_pagamento]
    trace_media = go.Bar(name='Média - {}'.format(forma_pagamento), x=df_forma_pagamento['produto'], y=df_forma_pagamento['mean'])
    trace_max = go.Bar(name='Máximo - {}'.format(forma_pagamento), x=df_forma_pagamento['produto'], y=df_forma_pagamento['max'])
    trace_min = go.Bar(name='Mínimo - {}'.format(forma_pagamento), x=df_forma_pagamento['produto'], y=df_forma_pagamento['min'])
    fig2.add_trace(trace_media)
    fig2.add_trace(trace_max)
    fig2.add_trace(trace_min)
fig2.update_layout(title='Estatísticas de Dias de Atraso por Produto e Forma de Pagamento', barmode='group')
fig1.update_layout(template='plotly_dark')
fig2.update_layout(template='plotly_dark')

fig1.show()
fig2.show()

```
